```python
### Fixing the Issues

# 1) Missing product prices
print("Missing product prices:", products['price'].isnull().sum())

# 2) Convert price to numeric
sale_items['price'] = pd.to_numeric(sale_items['price'], errors='coerce')
print("Missing sale_items price after conversion:", sale_items['price'].isnull().sum())

# 3) Negative sales
print("Negative sales rows:")
display(sales[sales['total_amount'] < 0])

# 4) Missing unit cost
print("Missing unit_cost:", sale_items['unit_cost'].isnull().sum())

# 5) Boxplot for outliers
plt.figure()
sns.boxplot(x=sale_items['total'])
plt.title("Distribution of Sale Item Totals")
plt.show()
```

```python

### FINAL DATA CLEANING


### 1 Remove negative sales (likely refunds or corrections)
sales = sales[sales['total_amount'] >= 0].copy()


### 2 Convert sale_items price to numeric (handle corrupted values)
sale_items['price'] = pd.to_numeric(sale_items['price'], errors='coerce')

### Drop rows where price could not be converted (only 1 row expected)
sale_items = sale_items.dropna(subset=['price']).copy()


### 3 Merge sale_items with products (to get cost & metadata)
df = sale_items.merge(products,
                      left_on='product_code',
                      right_on='code',
                      how='left')


### 4 Merge with sales table (to get sale-level info like date)
df = df.merge(sales,
              left_on='sale_id',
              right_on='id',
              how='left')


### 5 Handle missing unit_cost
### Fill missing unit_cost using cost from products table
df['unit_cost'] = df['unit_cost'].fillna(df['cost'])


### 6 Feature Engineering: Revenue & Profit
df['revenue'] = df['total']
df['profit'] = df['revenue'] - df['unit_cost']
df['margin_pct'] = (df['profit'] / df['revenue']) * 100


### 7 Optional: Investigate outliers (DO NOT auto-remove)
Q1 = df['revenue'].quantile(0.25)
Q3 = df['revenue'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR

high_value_transactions = df[df['revenue'] > upper_bound]

print("High value transactions (investigate, not removed):")
print(high_value_transactions.head())


### 8 Final Data Check
print("Final dataset shape:", df.shape)
print("\nMissing values:\n", df.isnull().sum())
```

```python

### HANDLE REMAINING MISSING VALUES

print("Remaining missing values before final fixes:")
print(df.isnull().sum())


### 1 Fix remaining missing unit_cost (2 rows)

# Fill missing unit_cost using product-level average cost
df['unit_cost'] = df.groupby('product_code')['unit_cost'].transform(
    lambda x: x.fillna(x.mean())
)

# If still missing (edge case), fill with overall median
df['unit_cost'] = df['unit_cost'].fillna(df['unit_cost'].median())


### 2 Recalculate profit & margin after fixing unit_cost
df['profit'] = df['revenue'] - df['unit_cost']
df['margin_pct'] = (df['profit'] / df['revenue']) * 100


### 3 Drop rows with missing product metadata (11 unmatched products)
df = df.dropna(subset=['label', 'type']).copy()


print("\nFinal clean dataset shape:", df.shape)
print("\nRemaining missing values after fixes:")
print(df.isnull().sum())
```

```python

### ADD DATETIME FEATURES

df['created_at'] = pd.to_datetime(df['created_at_y'])

df['hour'] = df['created_at'].dt.hour
df['month'] = df['created_at'].dt.month
df['day_of_week'] = df['created_at'].dt.day_name()
df['day'] = df['created_at'].dt.day
df['week'] = df['created_at'].dt.isocalendar().week
```

```python

### BUSINESS INTELLIGENCE ANALYSIS


import pandas as pd
import numpy as np
from datetime import datetime

print("="*70)
print("BUSINESS INTELLIGENCE FINDINGS")
print("="*70)
print(f"Analysis Date: {datetime.now().strftime('%Y-%m-%d %H:%M')}")
print(f"Total Transactions: {df['sale_id'].nunique():,}")
print(f"Total Products Sold: {df['product_code'].nunique():,}")

total_revenue = df['revenue'].sum()
total_profit = df['profit'].sum()
overall_margin = (total_profit / total_revenue * 100).round(1)

print(f"Total Revenue: ${total_revenue:,.2f}")
print(f"Total Profit: ${total_profit:,.2f}")
print(f"Overall Margin: {overall_margin}%")
print("="*70)


# ==========================================================
# Q1: Pareto Analysis (80% Revenue Rule)

print("\n Q1: REVENUE CONCENTRATION (80% RULE)")
print("-"*60)

product_revenue = (
    df.groupby(['product_code', 'label'])['revenue']
      .sum()
      .sort_values(ascending=False)
      .reset_index()
)

product_revenue['revenue_pct'] = (product_revenue['revenue'] / total_revenue * 100).round(2)
product_revenue['cumulative_pct'] = (
    product_revenue['revenue'].cumsum() / total_revenue * 100
).round(2)

# Identify products contributing to ~80%
cutoff_index = product_revenue[product_revenue['cumulative_pct'] >= 80].index[0]
products_80 = product_revenue.iloc[:cutoff_index+1]
pareto_count = len(products_80)
pareto_pct = round((pareto_count / len(product_revenue) * 100), 1)

print(f"\n PARETO INSIGHT:")
print(f"    {pareto_count} products ({pareto_pct}% of {len(product_revenue)} total SKUs)")
print(f"    Generate ${products_80['revenue'].sum():,.2f} ({products_80['cumulative_pct'].iloc[-1]}% of revenue)")
print(f"    First ${products_80['revenue'].iloc[0]:,.0f} product alone contributes {products_80['revenue_pct'].iloc[0]}%")

print("\n TOP 10 PRODUCTS BY REVENUE:")
print("-" * 80)
print(f"{'Rank':<4} {'Product':<35} {'Revenue':>12} {'% of Total':>10} {'Cumulative':>12}")
print("-" * 80)

for idx, row in product_revenue.head(10).iterrows():
    rank = idx + 1
    product_name = row['label'][:32] + '...' if len(row['label']) > 32 else row['label']
    print(f"{rank:<4} {product_name:<35} ${row['revenue']:>11,.2f} {row['revenue_pct']:>9.1f}% {row['cumulative_pct']:>11.1f}%")


# ==========================================================
# Q2: Peak Sales Hours

print("\n Q2: PEAK SALES HOURS ANALYSIS")
print("-"*60)

hourly_sales = (
    df.groupby('hour')['revenue']
      .agg(['sum', 'count', 'mean'])
      .rename(columns={'sum': 'total_revenue', 'count': 'transactions', 'mean': 'avg_ticket'})
      .sort_values('total_revenue', ascending=False)
)

# Add hour labels
hourly_sales['time_range'] = hourly_sales.index.map(lambda x: f"{x:02d}:00-{x+1:02d}:00")

peak_hour = hourly_sales.index[0]
peak_hour_revenue = hourly_sales.iloc[0]['total_revenue']
peak_hour_transactions = hourly_sales.iloc[0]['transactions']
peak_hour_avg = hourly_sales.iloc[0]['avg_ticket']

print(f"\n PEAK HOUR: {peak_hour:02d}:00 - {peak_hour+1:02d}:00")
print(f"    Revenue: ${peak_hour_revenue:,.2f}")
print(f"    Transactions: {peak_hour_transactions:,}")
print(f"    Avg Ticket: ${peak_hour_avg:,.2f}")

print("\n  TOP 5 PEAK HOURS:")
print("-" * 60)
print(f"{'Hour':<15} {'Revenue':>12} {'Transactions':>12} {'Avg Ticket':>12}")
print("-" * 60)

for hour, row in hourly_sales.head(5).iterrows():
    print(f"{row['time_range']:<15} ${row['total_revenue']:>11,.2f} {row['transactions']:>12,} ${row['avg_ticket']:>11,.2f}")

# Off-peak insight
off_peak_hours = hourly_sales.tail(5)
print(f"\n OFF-PEAK HOURS (Lowest 5):")
for hour, row in off_peak_hours.iterrows():
    print(f"   {row['time_range']:<15} ${row['total_revenue']:>11,.2f}")


# ==========================================================
# Q3: Highest Profit Margin Products (Weighted)

print("\n Q3: MOST PROFITABLE PRODUCTS (Weighted Margin)")
print("-"*60)

product_profit = (
    df.groupby(['product_code', 'label'])
      .agg({
          'revenue': 'sum',
          'profit': 'sum',
          'sale_id': 'count'  # transaction count
      })
      .rename(columns={'sale_id': 'transaction_count'})
)

product_profit['margin_pct'] = (
    product_profit['profit'] / product_profit['revenue'] * 100
).round(1)

product_profit['revenue_pct'] = (
    product_profit['revenue'] / total_revenue * 100
).round(1)

# Filter for meaningful volume (top 50% by revenue and min transactions)
revenue_threshold = product_profit['revenue'].quantile(0.5)
transaction_threshold = 5  # At least 5 transactions

product_profit_filtered = product_profit[
    (product_profit['revenue'] >= revenue_threshold) &
    (product_profit['transaction_count'] >= transaction_threshold)
]

top_margin_products = (
    product_profit_filtered
      .sort_values('margin_pct', ascending=False)
      .head(10)
)

print(f"\n ANALYSIS PARAMETERS:")
print(f"    Min revenue threshold: ${revenue_threshold:,.2f} (top 50%)")
print(f"    Min transactions: {transaction_threshold}")
print(f"    Products analyzed: {len(product_profit_filtered)} of {len(product_profit)} total")

print("\n TOP 10 HIGH-MARGIN PRODUCTS:")
print("-" * 85)
print(f"{'Rank':<4} {'Product':<35} {'Margin':>8} {'Revenue':>12} {'% of Total':>10} {'Transactions':>12}")
print("-" * 85)

for idx, (product, row) in enumerate(top_margin_products.iterrows(), 1):
    product_name = row.name[1] if isinstance(row.name, tuple) else row.name
    product_name = product_name[:32] + '...' if len(product_name) > 32 else product_name
    print(f"{idx:<4} {product_name:<35} {row['margin_pct']:>7.1f}% ${row['revenue']:>11,.2f} {row['revenue_pct']:>9.1f}% {row['transaction_count']:>12,}")

# Compare with average
avg_margin_all = (product_profit['profit'].sum() / product_profit['revenue'].sum() * 100).round(1)
avg_margin_filtered = (top_margin_products['profit'].sum() / top_margin_products['revenue'].sum() * 100).round(1)

print(f"\n MARGIN COMPARISON:")
print(f"    Overall average margin: {avg_margin_all}%")
print(f"    Top 10 products avg margin: {avg_margin_filtered}%")
print(f"    Premium over average: +{avg_margin_filtered - avg_margin_all:.1f}%")


# ==========================================================
# Q4: Peak Months & Seasonal Performance

print("\n Q4: MONTHLY PERFORMANCE ANALYSIS")
print("-"*60)

monthly_sales = (
    df.groupby('month')['revenue']
      .agg(['sum', 'count', 'mean'])
      .rename(columns={'sum': 'total_revenue', 'count': 'transactions', 'mean': 'avg_ticket'})
      .sort_values('total_revenue', ascending=False)
)

# Add month names
month_names = {
    1: 'January', 2: 'February', 3: 'March', 4: 'April',
    5: 'May', 6: 'June', 7: 'July', 8: 'August',
    9: 'September', 10: 'October', 11: 'November', 12: 'December'
}
monthly_sales['month_name'] = monthly_sales.index.map(month_names)

print("\n MONTHLY RANKING:")
print("-" * 70)
print(f"{'Month':<12} {'Revenue':>12} {'% of Year':>10} {'Transactions':>12} {'Avg Ticket':>12}")
print("-" * 70)

for month, row in monthly_sales.iterrows():
    revenue_pct = (row['total_revenue'] / total_revenue * 100).round(1)
    print(f"{row['month_name']:<12} ${row['total_revenue']:>11,.2f} {revenue_pct:>9.1f}% {row['transactions']:>12,} ${row['avg_ticket']:>11,.2f}")

peak_month = monthly_sales.index[0]
peak_month_name = month_names[peak_month]
peak_month_revenue = monthly_sales.loc[peak_month, 'total_revenue']
avg_monthly_revenue = monthly_sales['total_revenue'].mean()

print(f"\n PEAK MONTH: {peak_month_name}")
print(f"    Revenue: ${peak_month_revenue:,.2f}")
print(f"    {((peak_month_revenue/avg_monthly_revenue - 1) * 100):.0f}% above monthly average")
print(f"    Transactions: {monthly_sales.loc[peak_month, 'transactions']:,}")
print(f"    Avg Ticket: ${monthly_sales.loc[peak_month, 'avg_ticket']:,.2f}")

print(f"\n TOP 5 PRODUCTS IN {peak_month_name.upper()}:")
print("-" * 60)
print(f"{'Rank':<4} {'Product':<35} {'Revenue':>12} {'% of Month':>10}")
print("-" * 60)

month_total = monthly_sales.loc[peak_month, 'total_revenue']
peak_products = (
    df[df['month'] == peak_month]
      .groupby(['product_code', 'label'])['revenue']
      .sum()
      .sort_values(ascending=False)
      .head(5)
)

for idx, (product, revenue) in enumerate(peak_products.items(), 1):
    product_name = product[1] if isinstance(product, tuple) else product
    product_name = product_name[:32] + '...' if len(product_name) > 32 else product_name
    month_pct = (revenue / month_total * 100).round(1)
    print(f"{idx:<4} {product_name:<35} ${revenue:>11,.2f} {month_pct:>9.1f}%")


# EXECUTIVE SUMMARY

print("\n" + "="*70)
print(" EXECUTIVE SUMMARY")
print("="*70)

print(f"""
KEY FINDINGS:

1 PRODUCT CONCENTRATION:
    {pareto_count} products drive 80% of revenue (Pareto principle confirmed)
    Top product contributes {product_revenue['revenue_pct'].iloc[0]}% of total revenue
    Action: Focus inventory and marketing on these key products

2 PEAK OPERATING HOURS:
    Peak hour: {peak_hour:02d}:00-{peak_hour+1:02d}:00 (${peak_hour_revenue:,.2f})
    Peak hour transactions: {peak_hour_transactions:,}
    Action: Optimize staffing and promotions during peak windows

3 PROFITABILITY INSIGHTS:
    Overall margin: {overall_margin}%
    Top margin products average: {avg_margin_filtered}%
    Premium products deliver +{avg_margin_filtered - overall_margin:.1f}% margin
    Action: Promote and bundle high-margin products

4 SEASONAL PATTERNS:
    Peak month: {peak_month_name} (${peak_month_revenue:,.2f})
    {peak_month_name} is {((peak_month_revenue/avg_monthly_revenue - 1) * 100):.0f}% above average
    Action: Plan inventory and campaigns around seasonal peaks
""")
```

```python

### EXPORT CLEANED ORIGINAL DATASETS


import pandas as pd
from google.colab import files
import os
from datetime import datetime

# Create a folder for exports
export_folder = '/content/cleaned_original_datasets'
os.makedirs(export_folder, exist_ok=True)

print("="*70)
print(" EXPORTING CLEANED ORIGINAL DATASETS")
print("="*70)
print(f"Export Date: {datetime.now().strftime('%Y-%m-%d %H:%M')}")
print("="*70)


# 1. Export Cleaned Products Dataset

print("\n1 Exporting CLEANED PRODUCTS dataset...")

# Create a cleaned products dataframe
products_cleaned = products.copy()

# Fill missing prices (if any remain) - using median by type as before
products_cleaned['price'] = products_cleaned.groupby('type')['price'].transform(
    lambda x: x.fillna(x.median())
)
products_cleaned['price'] = products_cleaned['price'].fillna(products_cleaned['price'].median())

# Ensure no negative costs/prices
products_cleaned['cost'] = products_cleaned['cost'].abs()
products_cleaned['price'] = products_cleaned['price'].abs()

# Add calculated fields
products_cleaned['potential_margin'] = ((products_cleaned['price'] - products_cleaned['cost']) / products_cleaned['price'] * 100).round(2)
products_cleaned['price_to_cost_ratio'] = (products_cleaned['price'] / products_cleaned['cost']).round(2)

# Clean product labels (remove extra spaces, proper casing)
products_cleaned['label'] = products_cleaned['label'].str.strip().str.title()
products_cleaned['type'] = products_cleaned['type'].str.strip().str.lower()

# Export to CSV
products_path = f'{export_folder}/cleaned_products.csv'
products_cleaned.to_csv(products_path, index=False)

print(f"    Saved: {products_path}")
print(f"    Shape: {products_cleaned.shape}")
print(f"    Columns: {list(products_cleaned.columns)}")
print(f"    Missing values: {products_cleaned.isnull().sum().sum()}")
print(f"    Total inventory value: ${(products_cleaned['cost'] * 10).sum():,.2f} (estimated)")



# 2. Export Cleaned Sales Dataset

print("\n2 Exporting CLEANED SALES dataset...")

# Create a cleaned sales dataframe
sales_cleaned = sales.copy()

# Remove any remaining negative values (already done but double-check)
sales_cleaned = sales_cleaned[sales_cleaned['total_amount'] >= 0].copy()

# Convert created_at to datetime
sales_cleaned['created_at'] = pd.to_datetime(sales_cleaned['created_at'])

# Add date-based columns for easier analysis
sales_cleaned['date'] = sales_cleaned['created_at'].dt.date
sales_cleaned['year'] = sales_cleaned['created_at'].dt.year
sales_cleaned['month'] = sales_cleaned['created_at'].dt.month
sales_cleaned['day'] = sales_cleaned['created_at'].dt.day
sales_cleaned['hour'] = sales_cleaned['created_at'].dt.hour
sales_cleaned['day_of_week'] = sales_cleaned['created_at'].dt.day_name()
sales_cleaned['week'] = sales_cleaned['created_at'].dt.isocalendar().week

# Add calculated fields
sales_cleaned['payment_efficiency'] = (sales_cleaned['amount_paid'] / sales_cleaned['total_amount'] * 100).round(2)
sales_cleaned['discount_rate'] = (sales_cleaned['discount'] / sales_cleaned['total_amount'] * 100).round(2).fillna(0)

# Export to CSV
sales_path = f'{export_folder}/cleaned_sales.csv'
sales_cleaned.to_csv(sales_path, index=False)

print(f"    Saved: {sales_path}")
print(f"    Shape: {sales_cleaned.shape}")
print(f"    Columns: {list(sales_cleaned.columns)}")
print(f"    Missing values: {sales_cleaned.isnull().sum().sum()}")
print(f"    Total sales amount: ${sales_cleaned['total_amount'].sum():,.2f}")
print(f"    Date range: {sales_cleaned['date'].min()} to {sales_cleaned['date'].max()}")



# 3. Export Cleaned Sale Items Dataset

print("\n3 Exporting CLEANED SALE ITEMS dataset...")

# Create a cleaned sale_items dataframe
sale_items_cleaned = sale_items.copy()

# Ensure price is numeric (already done)
sale_items_cleaned['price'] = pd.to_numeric(sale_items_cleaned['price'], errors='coerce')

# Fill missing unit_cost using products cost (as per your logic)
# First, create a product cost mapping
product_cost_map = products.set_index('code')['cost'].to_dict()
sale_items_cleaned['unit_cost'] = sale_items_cleaned['unit_cost'].fillna(
    sale_items_cleaned['product_code'].map(product_cost_map)
)

# For any remaining missing unit_cost, use median by product code
sale_items_cleaned['unit_cost'] = sale_items_cleaned.groupby('product_code')['unit_cost'].transform(
    lambda x: x.fillna(x.median())
)

# Final fallback to overall median
sale_items_cleaned['unit_cost'] = sale_items_cleaned['unit_cost'].fillna(sale_items_cleaned['unit_cost'].median())

# Add calculated fields
sale_items_cleaned['profit'] = sale_items_cleaned['total'] - sale_items_cleaned['unit_cost']
sale_items_cleaned['margin'] = (sale_items_cleaned['profit'] / sale_items_cleaned['total'] * 100).round(2)
sale_items_cleaned['discount_percentage'] = (sale_items_cleaned['discount'] / sale_items_cleaned['price'] * 100).round(2)

# Convert created_at to datetime
sale_items_cleaned['created_at'] = pd.to_datetime(sale_items_cleaned['created_at'])
sale_items_cleaned['date'] = sale_items_cleaned['created_at'].dt.date
sale_items_cleaned['hour'] = sale_items_cleaned['created_at'].dt.hour

# Export to CSV
sale_items_path = f'{export_folder}/cleaned_sale_items.csv'
sale_items_cleaned.to_csv(sale_items_path, index=False)

print(f"    Saved: {sale_items_path}")
print(f"    Shape: {sale_items_cleaned.shape}")
print(f"    Columns: {list(sale_items_cleaned.columns)}")
print(f"    Missing values: {sale_items_cleaned.isnull().sum().sum()}")
print(f"    Total revenue: ${sale_items_cleaned['total'].sum():,.2f}")
print(f"    Total profit: ${sale_items_cleaned['profit'].sum():,.2f}")
print(f"    Unique products: {sale_items_cleaned['product_code'].nunique()}")



# 4. Create a data dictionary / README

print("\n4 Creating data dictionary...")

readme_content = f"""
# CLEANED DATASETS - DATA DICTIONARY
Generated: {datetime.now().strftime('%Y-%m-%d %H:%M')}

## 1. cleaned_products.csv
- id: Unique product identifier
- code: Product code (matches sale_items.product_code)
- label: Product name (cleaned, title case)
- type: Product type/category (lowercase)
- cost: Unit cost to business
- price: Selling price (missing values filled)
- potential_margin: Calculated margin percentage
- price_to_cost_ratio: Price divided by cost

## 2. cleaned_sales.csv
- id: Unique sale identifier
- sale_ref: Sale reference number
- total_amount: Total sale amount
- amount_given: Amount given by customer
- amount_paid: Amount actually paid
- amount_due: Amount still due
- discount: Discount applied
- created_at: Timestamp of sale
- date: Sale date
- year/month/day/hour: Extracted time components
- day_of_week: Day name
- week: Week number
- payment_efficiency: (amount_paid/total_amount)*100
- discount_rate: (discount/total_amount)*100

## 3. cleaned_sale_items.csv
- id: Unique line item identifier
- sale_id: References sales.id
- product_code: References products.code
- price: Price at time of sale
- discount: Discount on this item
- total: Final amount after discount
- unit_cost: Cost of product (missing values filled)
- created_at: Timestamp
- profit: total - unit_cost
- margin: (profit/total)*100
- discount_percentage: (discount/price)*100
- date/hour: Extracted time components

## DATA QUALITY NOTES:
- All negative sales values removed
- All prices converted to numeric
- Missing unit_cost filled using products.cost
- Product labels cleaned and standardized
- Date fields properly formatted as datetime
- No remaining null values in key fields
"""

readme_path = f'{export_folder}/README.txt'
with open(readme_path, 'w') as f:
    f.write(readme_content)

print(f"    Saved: {readme_path}")



# Create a zip file for easy download

print("\n" + "="*70)
print(" CREATING ZIP FILE FOR DOWNLOAD")
print("="*70)

import zipfile

zip_path = '/content/cleaned_original_datasets.zip'
with zipfile.ZipFile(zip_path, 'w') as zipf:
    for file in os.listdir(export_folder):
        file_path = os.path.join(export_folder, file)
        zipf.write(file_path, arcname=file)

print(f" Zip file created: {zip_path}")
print(f"\n FILES INCLUDED:")
print("-" * 40)

total_size = 0
for file in os.listdir(export_folder):
    file_size = os.path.getsize(os.path.join(export_folder, file)) / 1024  # KB
    total_size += file_size
    print(f"    {file} ({file_size:.1f} KB)")

print("-" * 40)
print(f"   Total size: {total_size:.1f} KB ({total_size/1024:.2f} MB)")


# ==========================================================
# Preview the first few rows of each cleaned dataset
# ==========================================================
print("\n" + "="*70)
print("PREVIEW OF CLEANED DATASETS")
print("="*70)

print("\n cleaned_products.csv (first 3 rows):")
print("-" * 40)
print(products_cleaned[['code', 'label', 'type', 'cost', 'price', 'potential_margin']].head(3).to_string())

print("\n cleaned_sales.csv (first 3 rows):")
print("-" * 40)
print(sales_cleaned[['id', 'total_amount', 'discount', 'date', 'day_of_week', 'payment_efficiency']].head(3).to_string())

print("\n cleaned_sale_items.csv (first 3 rows):")
print("-" * 40)
print(sale_items_cleaned[['sale_id', 'product_code', 'total', 'profit', 'margin', 'date']].head(3).to_string())


# ==========================================================
# Download the zip file
# ==========================================================
print("\n" + "="*70)
print(" DOWNLOADING FILES")
print("="*70)
print("\nClick the link below to download all cleaned CSV files:")

from google.colab import files
files.download(zip_path)

print("\n Download initiated!")
print("\n Your cleaned original datasets are ready for Power BI!")
print("   Files downloaded: cleaned_products.csv, cleaned_sales.csv, cleaned_sale_items.csv")
print("   Plus: README.txt with data dictionary")
```
