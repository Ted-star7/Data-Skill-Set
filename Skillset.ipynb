```python
### Issues in the Data SET
### 1)Missing Prices in Products - total rows = 509-412 non-null So:97 missing
### 2)Price Column in sale_items is OBJECT, should be numeric
### 3)Missing unit_cost - Total rows = 8004-7507 non-null So 497 missing unit_cost values
### 4)Negative Sales Values- -4788.414
### 5)Extreme Outliers- max total = 991,584 while median is 800
```

```python
### Fixing the Issues

# 1) Missing product prices
print("Missing product prices:", products['price'].isnull().sum())

# 2) Convert price to numeric
sale_items['price'] = pd.to_numeric(sale_items['price'], errors='coerce')
print("Missing sale_items price after conversion:", sale_items['price'].isnull().sum())

# 3) Negative sales
print("Negative sales rows:")
display(sales[sales['total_amount'] < 0])

# 4) Missing unit cost
print("Missing unit_cost:", sale_items['unit_cost'].isnull().sum())

# 5) Boxplot for outliers
plt.figure()
sns.boxplot(x=sale_items['total'])
plt.title("Distribution of Sale Item Totals")
plt.show()
```
